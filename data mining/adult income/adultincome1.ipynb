```python
a_income = pd.get_dummies(a_income, columns=['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race','sex', 'native.country'], drop_first=True)
```

```python
a_income.head()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
scaler = StandardScaler()
numerical_features = ['age', 'fnlwgt', 'capital.gain', 'capital.loss', 'hours.per.week']
a_income[numerical_features] = scaler.fit_transform(a_income[numerical_features])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
a_income['income'] = le.fit_transform(a_income['income'])
```

```python
from sklearn.model_selection import train_test_split
x = a_income.drop('income', axis=1)
y = a_income['income']
x_test, x_train, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=42)
```
